# What is the unemployment v/s employment rate at Kerala?

```python
#Employment rate at Kerala

df['area']=df['Region'].str.contains('Kerala', case=False).sum()
df.groupby('area')[' Estimated Employed'].sum()
```

```python
#Unemployment rate at Kerala

df.groupby('area')[' Estimated Unemployment Rate (%)'].sum()
```
